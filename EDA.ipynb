2.4.1. Visualizing Categorical Data

A histogram, a barchart or a categorical plot, helps us to visual different values and representation of data distribution. Histogram can be used to identify the central tendency, dispersion, and skewness of data. Each column presents each characteristics and patterns in the data. Therefore, users can analyze, predict trend and detect outliers or highlights in the dataset.

```python
from sklearn.preprocessing import LabelEncoder , StandardScaler
import pandas as pd
df = pd.read_csv('Obesity_raw.csv')
# Initialize a LabelEncoder
encoder = LabelEncoder()
# Encode all categorical columns
for column in df.select_dtypes(include=['object']).columns:
    df[column] = encoder.fit_transform(df[column])

# Compute the correlation matrix
correlation_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Obesity_raw.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Convert categorical variables to numeric codes
df['Gender_Code'] = df['Gender'].astype('category').cat.codes
df['NObeyesdad_Code'] = df['NObeyesdad'].astype('category').cat.codes

# Select relevant variables for the scatterplot matrix
scatter_data = df[['Gender_Code', 'NObeyesdad_Code']]

# Create scatterplot matrix
sns.pairplot(scatter_data, diag_kind='kde', corner=True, plot_kws={'alpha': 0.6, 's': 50})
plt.suptitle('Scatterplot Matrix: Gender and NObeyesdad', y=1.02, fontsize=16)
plt.show()
```
